```python
# transform data in the alchemy_keywords column into a format that can be parsed by vectorizers
keywords = []

for line in train_data['alchemy_keywords']:
    if(not pd.isnull(line)):
        words_list = []
        try:
            for word in line.split():
              try:
                encoded_word = word.encode('-utf-8')
              except UnicodeDecodeError:
                encoded_word = ""
              words_list.append(encoded_word)
        except AttributeError:
            continue  
        keywords_str = " ".join(str(x) for x in words_list)
    else:
        keywords_str = ""
    keywords.append(keywords_str)

keyword_series = pd.Series(keywords)
keyword_series[80:100]
```

```python
keyword_features = pd.DataFrame(vectorizer.fit_transform(keyword_series).toarray())
keyword_features.shape
```

```python
# concatenate all processed features into one big features matrix
features = pd.concat([numerical_data, alchemy_category_features, keyword_features], axis=1)
```

```python
# Get rid of NaN values using imputer. 
# Employing the "most_frequent" strategy, but this will probably make certain binary features useless 
# (ex: is_news has many NaNs, replacing those values with 1 probably isn't the best approach)
imputer.fit(features)
imputed_features = pd.DataFrame(imputer.transform(features))
imputed_features
```

```python
X = imputed_features
y = train_data.label
```
